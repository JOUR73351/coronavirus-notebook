```python
c = Census("53bc2659c51f37cee0b647581299137dd5ab602d")
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/nychealth/coronavirus-data/master/tests-by-zcta.csv', dtype={'MODZCTA': 'str'})
df = df[1:]
df.head()
```

```python
zip_codes = df['MODZCTA'].tolist()
```

```python
categories = ['B01001_001E']
```

```python
all_population = c.acs5.zipcode(categories, Census.ALL)
all_population
```

```python
population = pd.DataFrame(all_population)
population['B01001_001E'] = population['B01001_001E'].astype('int32')
population.info()
```

```python
merged = df.merge(population, left_on='MODZCTA', right_on='zip code tabulation area')
merged.info()
```

```python
merged['per100000'] = merged['Positive']*100000 / merged['B01001_001E']
merged.head()
```
