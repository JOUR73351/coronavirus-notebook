```python
import pandas as pd
import requests
pd.options.display.max_rows = 335
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-counties.csv', dtype={'fips': str})
df.head()
```

```python
df['joined'] = df['county'] + ", " + df['state']
df.head()
```

```python
most_recent_date = df.date.max()
most_recent_date
```

```python
most_recent = df[df.date == most_recent_date]
most_recent.shape
```

```python
most_recent.sort_values("cases", ascending=False)
```

```python
most_recent[most_recent['fips'].isnull()]
```

```python
centroids = pd.read_csv('../data/centroids.csv')
centroids = centroids[['joined', 'lat', 'lng']]
centroids.head()
```

```python
states = most_recent.merge(centroids, on="joined")
states
```

```python
geocoded = pd.read_csv('../data/geocoded.csv', dtype={'STATEFP': str, 'COUNTYFP': str})
geocoded.head()
```

```python
geocoded['fips'] = geocoded['STATEFP'] + geocoded['COUNTYFP']
geocoded.head()
```

```python
tomerge = geocoded[['fips', 'lat', 'lng']]
tomerge
```

```python
merged = most_recent.merge(tomerge, on='fips')
merged
```

```python
merged = merged.append(states)
merged
```

```python
merged.to_csv('../output/counties.csv', index=False)
```
