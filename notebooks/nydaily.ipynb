```python
import pandas as pd
```

```python
states = pd.read_csv('https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-states.csv')
states
```

```python
ny = states[states['state'] == 'New York']
ny
```

```python
ny['daily_cases'] = ny['cases'].diff()
ny['daily_deaths'] = ny['deaths'].diff()
ny = ny.fillna(0)
```

```python
ny.to_csv('../output/ny.csv', index=False)
```
